```python
import pandas as pd
import numpy as np
from category_encoders import TargetEncoder
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("train.csv")
df = df.drop(['POSTED_BY'],axis=1)
x_train, x_test, y_train, y_test = train_test_split(df.drop(['TARGET(PRICE_IN_LACS)'],axis=1),df['TARGET(PRICE_IN_LACS)'],test_size=0.25,random_state=42)
x_train.shape,x_test.shape


print(x_train.shape,x_test.shape)
x_train.head()
```

```python
x_train['SQUARE_FT'] = x_train['SQUARE_FT'] / x_train['SQUARE_FT'].max()
x_train['LONGITUDE'] = x_train['LONGITUDE'] / 180
x_train['LATITUDE'] = x_train['LATITUDE'] / 180

x_test['SQUARE_FT'] = x_test['SQUARE_FT'] / x_test['SQUARE_FT'].max()
x_test['LONGITUDE'] = x_test['LONGITUDE'] / 180
x_test['LATITUDE'] = x_test['LATITUDE'] / 180
```

```python
x_test.head()
```

```python
# x_train = x_train.drop(['TARGET(PRICE_IN_LACS)'],axis=1)
y_train = y_train / 1000

# x_test = x_test.drop(['TARGET(PRICE_IN_LACS)'],axis=1)
y_test = y_test / 1000

x_train.shape,y_train.shape
```

```python
encoder = TargetEncoder(cols=['BHK_OR_RK','ADDRESS'])
encoder.fit(x_train, y_train)
x_train_enc = encoder.transform(x_train)

encoder.fit(x_test, y_test)
x_test_enc = encoder.transform(x_test)

x_train_enc.head()
```

```python
y_train.describe()
```

```python
# Asumiendo que ya has importado las bibliotecas necesarias y definido tus datos x_train, y_train, x_test, y_test

# Definir el modelo
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(x_train.shape[1],), activation='sigmoid'),
    keras.layers.Dense(8, activation='swish'),
    keras.layers.Dense(1)
])

# Learning rate decay
lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.01,
    decay_steps=10000,
    decay_rate=0.9)

optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)

# Compilar el modelo
model.compile(
    optimizer=optimizer,
    loss=keras.losses.MeanSquaredError(),
    metrics=[keras.metrics.MeanAbsoluteError()]
)

```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)

# Entrenar el modelo y guardar el historial
history = model.fit(x_train_enc, y_train, epochs=100, validation_data=(x_test_enc, y_test))

# Graficar el "training loss" y el "validation loss"

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation/Test Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss Value')
plt.legend()
plt.show()
```

```python
model.evaluate(x_test_enc, y_test)
```
